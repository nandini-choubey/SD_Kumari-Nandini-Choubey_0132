```python
class Student:
    def __init__(self, name, marks, id):
        self.name = name
        self.id = id
        self.marks = marks

    def display(self):
        print("ID:" , self.id)
        print("Name:" , self.name)
        print("Marks:" , self.marks)

s1 = Student(101, "Alice", 95)
s1.display()
```
