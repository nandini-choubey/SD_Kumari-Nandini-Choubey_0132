```python
def test_fib(n):
  a,b = 0,1
  for i in range (n):
    yield a
    a,b = b , a+b
```

```python
for i in test_fib(7) :
  print (i)
```

```python
def test_fib1(n):
  a,b = 0,1
  while True :
    yield a
    a,b = b , a+b
```
